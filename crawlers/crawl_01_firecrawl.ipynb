```python
from firecrawl import FirecrawlApp
```

```python
firecrawl_key="xxx"
```

```python
app = FirecrawlApp(api_key=firecrawl_key)
```

```python
# Crawl a website
crawl_url = 'https://sr.se'
params = {
    'crawlerOptions': {
        'excludes': ['blog/*'],
        'includes': [], # leave empty for all pages
        'limit': 10,
    },
    'pageOptions': {
        'onlyMainContent': True
    }
}
```

```python
crawl_result = app.crawl_url(crawl_url, params=params, wait_until_done=True, timeout=15)
```

```python
# Get the markdown
for result in crawl_result:
    print(result['markdown'])
```
